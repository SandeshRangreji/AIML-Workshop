# Variables in Python



There is no need to "declare" a variable. A variable is created the moment you assign a value to it.
> variable_name = value

```python
i = 5
name = "John Doe"
print(i, name)
```

Moreover, Python is a *dynamically typed language*. This means that a variable can change its type during run time.

```python
x = 3 # x is of type int
print(x)
x = "This is a string." # x is now of type str
print(x)
```

### Type-Casting
If you want to specify the data type of the variable, you can do so by type-casting it.

```python
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print(x, y, z)
```

```python
print(type(x))
print(type(y))
print(type(z))
```

```python
# Multiple assignments
x, y, z = float(3), int(3), str(3)
print(x, y, z)
```

## Lists


### List Items
List items are:
<ul>
    <li> Ordered: have a defined order that will not change. New items will be placed at the end of the list</li>
    <li> Indexed: List items are indexed, the first item has index <code>[0]</code>, the second item has <code>[1]</code> etc. </li>
    <li> Changeable: We can change, add and remove items in a list after it has been created </li>
    <li> Duplicates: Since list items are indexed, list items can have same values. </li>
</ul>


```python
li = []
# add list items
li.append("a")
li.append("b")
li.append("c")
li.append("e")
print(li)
```

```python
# find length of a list
print( len(li))
```

```python
# indexing
print(li[0])
print(li[1])
print(li[-1])
```

```python
# changing an item
li[3] = "d"
print(li)
li[1] = 1
print(li)
```

```python
# removing items
li.remove("c")
print(li)
li.pop(0)
print(li)
```

```python
# joining lists
li1 = [1, 2]
li2 = [3, 4]
li = li1 + li2
print(li)
```

## Tuples
Tuple items are:
<ul>
    <li> Indexed </li>
    <li> Ordered </li>
    <li> Unchangeable </li>
    <li> Allow duplicates </li>
</ul>

```python
# initialize
tup = ("one", )
# NOTE: tup = ("one") is NOT a tuple
tup = ("one", "two", "three", "four")
# unpacking
a, b, c, d = tup
print(a, b, c, d)
```

## Sets
Set items are:
<ul>
    <li> Unordered </li>
    <li> Unchangable </li>
    <li> Duplicates <b>not</b> allowed </li>
</ul>

```python
# initializing
set1 = {"one", "two", "three", "four"}
print(len(set1))
```

```python
# adding to a set
set1.add("five")
print(set1)
# removing from a set
set1.remove("one")
print(set1)
```

```python
# no duplicates
li =  [1, 2, 3, 4, 5, 2, 4, 5, 1, 2, 3, 5, 6, 3, 3, 2, 1, 4, 3]
print(li)
set1 = set(li)
print(set1)
```

## Dictionaries
Dictionaries are:
<ul>
    <li> Changeable </li>
    <li> Ordered (from Python 3.7 ONLY) </li>
    <li> Duplicates <b>not</b> allowed </li>
</ul>

```python
dict1 = {
    "name" : "John",
    "age" : 36,
    "gender" : "Male",
    "graduate" : True,
    "children" : ["Joe", "Mary"]
}
print(dict1["name"])
# ERROR
# print(dict[name])
# print(dict["Name"])
print(dict1.get("Name", "default_val"))
print(dict1["children"][0])
print(len(dict1))
print(dict1)
```

```python
# keys and values
keys = dict1.keys()
values = dict1.values()
print(keys, values)
```

```python
# update
dict1["name"] = "Jack"
print(dict1)
dict1.update({"name" : "John", "age":37, "dob":"02/06/1985"})
print(dict1)
```

```python
# remove item
dict1.pop("gender")
print(dict1)
```

```python
# Arithmetic Operators
x, y = 5, 10
print("Addition: " + str(x+y) )
print("Subtraction: " + str(x-y) )
print("Multiplication: " + str(x*y) )
print("Division: " + str(x/y) )
print("Modulus: " + str(x%y) )
print("Exponentiation: " + str(x**y) )
print("Floor division: " + str(x//y) )
```

```python
# Assignment Operators
x += 2
y -= 9
print("x:" + str(x) + ", y:" + str(y))
```

```python
# Comparision Operators
print( x == y )
print( x > y )
print( x < y )
print( x != y )
print( 5 >= 6 )
print( 5 <= 5 )
```

```python
# Logical Operators
print( 1<2 and 3<4 )
print( 1<2 or 3<4)
print( (1<2 and 3>4) or 4<5 )
print( not 1<2)
```

```python
# Membership Operators
li = [ 1, 2, 3]
print( 2 in li)
print( 4 in li)
print( 1 not in li)
```

## For loops

```python
# print 0-10
for i in range(11):
    print(i)
```

```python
# print 5-10
for i in range(5,11):
    print(i)
```

```python
# print all odd numbers less than 10
for i in range(1, 11, 2):
    print(i)
```

```python
# looping through a list
li = ["one", "two", "three"]
for i in li:
    print(i)
```

```python
for key, value in dict1.items():
    print(key, value)
```

```python
# nested loop
for i in range(3):
    for j in range(3):
        print(i, j)
```

## While Loops

```python
i = 10
while i>5:
    print("condition met", i)
    i -= 1
print(i)
```

```python
i = 10
while True:
    if i>5:
        print("condition met", i)
        i -= 1
        continue
    else:
        i -= 1
        break
#     i=100
print(i)
```
